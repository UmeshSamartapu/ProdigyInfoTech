## Step 3: Load the Stable Diffusion Model

```python
from diffusers import StableDiffusionPipeline
import torch

# Load the model (no revision needed)
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16  # Enables FP16 usage for performance
)

pipe = pipe.to("cuda")  # Move to GPU
pipe.enable_attention_slicing()  # Reduce VRAM usage
```

## Step 4: Define prompts and generate images

```python
# Define a list of prompts
prompts = [
    "A futuristic cyberpunk city skyline at night with neon lights, 8k ultra detailed",
    "A forest made of crystal trees under a glowing purple sky, digital art",
    "A realistic photo of a cowboy riding a horse on Mars during sunset",
    "A steampunk airship flying over a medieval castle, concept art",
    "An astronaut penguin floating in space, cartoon style"
]

# Generate and save each image
for idx, prompt in enumerate(prompts):
    image = pipe(prompt, num_inference_steps=50).images[0]
    image.save(f"output_{idx+1}.png")
    print(f"✅ Generated: output_{idx+1}.png | Prompt: {prompt}")
```
